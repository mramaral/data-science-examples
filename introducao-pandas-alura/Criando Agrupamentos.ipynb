## Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
dados['Valor'].mean()
```

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca', 'Lins de Vasconcelos', 'Grajaú', 'Centro']
selecao = dados['Bairro'].isin(bairros)
dados2 = dados
dados = dados[selecao]
dados.head(10)
```

```python
dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
grupo_bairro2 = dados2.groupby('Bairro')
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print('{} --> {}'.format(bairro, data.Valor.mean()))
```

```python
grupo_bairro['Valor', 'Condominio'].mean().round(2)
```

## Estatisticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min': 'minimo', 'max': 'maximo', 'sum': 'soma'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = [20, 10])
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```

```python
fig2 = grupo_bairro2['Valor'].mean().plot.bar(color = 'blue')
fig2.set_ylabel('Valor do Aluguel')
fig2.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```
